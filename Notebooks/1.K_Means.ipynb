```python
# Find high correlations
correlation = df.corr()

plt.figure(figsize=(18, 14))
sns.heatmap(correlation, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap", fontsize=18)
plt.show()

# Threshold for high correlation
threshold = 0.8

# Find pairs of features with high correlation, avoiding redundancy
high_correlation = []

for col in correlation.columns:
    for row in correlation.index:
        if abs(correlation.loc[row, col]) > threshold and row != col and (col, row) not in high_correlation and (row, col) not in high_correlation:
            high_correlation.append((row, col, correlation.loc[row, col]))

# Print high correlations
if high_correlation:
    for item in high_correlation:
        print(f"Features: {item[0]} and {item[1]} have correlation of {item[2]:.2f}")
else:
    print("No feature pairs with high correlation found.")
```

This graph shows variance by PCA components. The horizontal line shows the number of components and teh vertical line shows the variance. We want to keep 90 to 95 percent of the information in the dataset so we're going to use 35 to 40 components.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

numeric_df = df_imputed.select_dtypes(include=[np.number])

# standardization
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

# PCA
pca = PCA(n_components=30)
pca_data = pca.fit_transform(scaled_data)

# change to dataframe
pca_df = pd.DataFrame(pca_data, columns=[f'PC{i+1}' for i in range(30)])

pca_df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

scores = []
range_n_clusters = range(2, 11)

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(pca_df)
    score = silhouette_score(pca_df, cluster_labels)
    scores.append(score)
    print(f"Clusters: {n_clusters} | Silhouette Score: {score:.4f}")
```

```python
plt.plot(range_n_clusters, scores, marker='o')
plt.title('Silhouette Score vs Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

# Unsupervised Feature Importance via Supervised Modeling
### Code Explanation
In the code cell below we first use KMeans clustering on the dataset with all the 50 features.
Then we use the lables that this model has assigned to each of the datapoints and add a new column to the data set called 'Cluster'.
We then use a random forest model and train it on this dataset with the new column. Needless to say, The cluster column is our target column. Then we use this model and obtain the importance of each feature and at the end sort them and print out the 10 most important features.
### Output Explanation
We see the 10 most important features. We had 50 features. If each of them had the same importance the number written infront of them should've been 0.02 meaning 2%. So if we compare that with the numbers written infront of these 10 features, it is appearant that these features have 2-3 times more importance than the average. So they are key in forming clusters.
### Psychological Interpretation
When we look at the 10 features, we see that they show openness and flexability (Openness to Cultures, Intellectual Curiosity) and also they are related to trustworthiness and team-work abilities (Trust, Reliability, Compassoin, Cooperation).

```python
from sklearn.cluster import KMeans
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

# Applying KMeans to df before PCA
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df_imputed)

# Add the cluster labels to the dataframe
df_imputed['Cluster'] = kmeans.labels_

# RandomForestClassifier to get feature importance
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(df_imputed.drop(columns=['Cluster']), df_imputed['Cluster'])

# Get feature importance
importances = rf.feature_importances_

# top important features
features = df_imputed.drop(columns=['Cluster']).columns
feature_importance = pd.Series(importances, index=features).sort_values(ascending=False)

# Display the top 10 most important features
print(feature_importance.head(10))
```

# KMeans with only important features

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

important_features = [
    'Artistic Interests', 'Trust in Others', 'Reliability', 'Cooperation',
    'Preference for Novelty', 'Intellectual Curiosity', 'Modesty',
    'Openness to Diverse Cultures', 'Compassion', 'Altruism'
]

X = df_imputed[important_features]
X_scaled = StandardScaler().fit_transform(X)

silhouette_scores = []
k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)
    print(f'Clusters: {k} | Silhouette Score: {score:.4f}')

plt.plot(k_values, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.grid(True)
plt.show()
```

# Evaluating KMeans with 3 metrics
### Describing the metrics
The metrics used to evaluate the clustering model is these three:
- Silhouette score: Is between -1 and 1 and shows how close each data point is to its own cluster and how far it is from other clusters.

- Davis-Bouldin Index (DBI): This metric is the mean of the distance between clusters in proportion to their inner diagonal. It's usually between 0.5 to 3. The lower the better.

- Calinski-Harabasz Index (CH): Doesn't have a range. The higher the better. It shows the variance between clusters relative to the variance inside the clusters.

### Output
From the cell of the output we can deduct that k = 2 has the best score in all the metrics. As we add to the number of clusters, all the three metrics become worse. Overall the quality of the clustering is not very good. It's either because the data points are not that different from each other and the data is homogeneous or linear clustering doesn't work very well.

```python
important_features = [
    'Artistic Interests', 'Trust in Others', 'Reliability', 'Cooperation',
    'Preference for Novelty', 'Intellectual Curiosity', 'Modesty',
    'Openness to Diverse Cultures', 'Compassion', 'Altruism'
]

df_filtered = df_imputed[important_features]
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# Range for number of clusters
k_range = range(2, 6)
scores = {}

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(df_filtered)

    silhouette_avg = silhouette_score(df_filtered, labels)
    db_index = davies_bouldin_score(df_filtered, labels)
    ch_score = calinski_harabasz_score(df_filtered, labels)

    scores[k] = {
        'Silhouette': silhouette_avg,
        'Davies-Bouldin': db_index,
        'Calinski-Harabasz': ch_score
    }

    print(f"Clusters: {k}")
    print(f"  Silhouette Score:       {silhouette_avg:.4f}")
    print(f"  Davies-Bouldin Index:   {db_index:.4f}  (lower is better)")
    print(f"  Calinski-Harabasz Score:{ch_score:.2f}  (higher is better)\n")
```

# Using Combinational Features

### Combinational Features
This time combinational features are being created:
- Social_Score: obtained from the mean of Trust in Others, Cooperation, Compation, Altruism
- Intellectual_Cultural_Score: obtained from the mean of Artistic Interests, Pereference for Novelty, Intellectual Curiosity, Openness to Diverse Cultures
- Personal_charachter_score: obtained from the mean of Reliability, Modesty

### Describing the output

From the cell of the output we can deduct that k = 2 has the best score in all the metrics. As we add to the number of clusters, all the three metrics become worse.
If we compare the output of this cell with the previous cell we see that the scores are overall better in this way. But not that much high.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import pandas as pd

# constructing combinational features
df_combined = pd.DataFrame()
df_combined['Social_Score'] = df_imputed[['Trust in Others', 'Cooperation', 'Compassion', 'Altruism']].mean(axis=1)
df_combined['Intellectual_Cultural_Score'] = df_imputed[['Artistic Interests', 'Preference for Novelty', 'Intellectual Curiosity', 'Openness to Diverse Cultures']].mean(axis=1)
df_combined['Personal_Character_Score'] = df_imputed[['Reliability', 'Modesty']].mean(axis=1)

# testing different 'k's
print("Testing different cluster numbers...")
best_score = -1
best_k = None

for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(df_combined)

    silhouette = silhouette_score(df_combined, cluster_labels)
    dbi = davies_bouldin_score(df_combined, cluster_labels)
    ch = calinski_harabasz_score(df_combined, cluster_labels)
    print(f"Clusters: {k}")
    print(f"  Silhouette Score:       {silhouette:.4f}")
    print(f"  Davies-Bouldin Index:   {dbi:.4f}")
    print(f"  Calinski-Harabasz Score:{ch:.2f} \n")
```
